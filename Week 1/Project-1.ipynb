```python
# Finding simple interest
p = int(input("Principal interest= "))
r = int(input("rate= "))
t = int(input("time= "))
A = p*(1 + (r/100)*t)
print("Simple interest= ", A)

# Finding compound interest
n= int(input("Number of compound years= "))
rate_over_num= r/n
c = p*(1 + pow(rate_over_num,(n*t)))
print("Compund interest =" , c)

# Finding the annuity plan
PMT = int(input("Periodic payment size=" ))
anuity = PMT * ((pow((1 + rate_over_num), (n*t)) - 1)/rate_over_num)

print("Annuity Plan= ", anuity)
```
